```python
import pandas as pd 
import spotipy 
sp = spotipy.Spotify() 
from spotipy.oauth2 import SpotifyClientCredentials 

#Premium
cid ="key1" 
secret = "key2" 

#Free
#cid ="key3" 
#secret = "key4" 

client_credentials_manager = SpotifyClientCredentials(client_id=cid, client_secret=secret) 
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager) 
sp.trace=False 

playlist = sp.user_playlist_tracks("kesarleon", "2tiye2096SXAS4uMnFsUFP") #Masterpiece in progress
#playlist = sp.user_playlist_tracks("22xma24ys4af6t67voqa263qa", "2HjfYBNh0j5ZqyPBrrn4us") #Las de la semana
#playlist = sp.user_playlist_tracks("22xma24ys4af6t67voqa263qa", "2aiSWacEwulc3IVj5Sl8L7") #Definitivamente me gustan

tracks = playlist['items']

while playlist['next']:
        playlist = sp.next(playlist)
        tracks.extend(playlist['items'])

ids = [] 
for i in range(len(tracks)): 
    ids.append(tracks[i]["track"]["id"]) 

features1=pd.DataFrame(sp.audio_features(ids[0:99]))
features2=pd.DataFrame(sp.audio_features(ids[100:199]))
features3=pd.DataFrame(sp.audio_features(ids[200:299]))
features4=pd.DataFrame(sp.audio_features(ids[300:399]))
features5=pd.DataFrame(sp.audio_features(ids[400:499]))
df=pd.concat([features1,features2,features3,features4,features5])


df
```
